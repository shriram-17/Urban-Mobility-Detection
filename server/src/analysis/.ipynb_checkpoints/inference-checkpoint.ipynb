```python
df.drop(columns=['congestion_level','timestamp'],inplace= True)
```

```python
from sklearn.cluster import KMeans
```

```python
df_encoded = pd.get_dummies(df, columns=['weather_conditions', 'road_type', 'vehicle_type'])
df_encoded
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_encoded)


kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_features)

df['cluster'] = kmeans.labels_

print("Clustered DataFrame:")
print(df)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce the dimensionality to 2 using PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame for the principal components
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Add the cluster labels to the PCA DataFrame
df_pca['cluster'] = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(8, 6))
for cluster_label in range(3):  # Adjust based on the number of clusters
    cluster_data = df_pca[df_pca['cluster'] == cluster_label]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=f'Cluster {cluster_label}')

plt.title('Clusters based on the first two principal components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = pd.read_csv("urban_mobility_data_past_year.csv",index_col='timestamp')
data.head()
```

```python
grouped_data = data.groupby('vehicle_type')['congestion_level'].mean().reset_index()

public_transport = ['Bus', 'Train']
grouped_data['vehicle_category'] = grouped_data['vehicle_type'].apply(lambda x: 'Public' if x in public_transport else 'Private')

agg_data = grouped_data.groupby('vehicle_category')['congestion_level'].mean()
plt.figure(figsize=(8, 6))
agg_data.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Congestion Levels by Vehicle Category')
plt.xlabel('Vehicle Category')
plt.ylabel('Average Congestion Level')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

df_encoded = pd.get_dummies(data_new, columns=['weather_conditions', 'road_type', 'vehicle_type'])

X = df_encoded.drop(columns=['congestion_level', 'timestamp'])
y = df_encoded['congestion_level']
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(),
    "Support Vector Machine": SVR(),
    "K-Nearest Neighbors": KNeighborsRegressor()
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

results = {}

for name, model in models.items():
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict y_train
    y_train_pred = model.predict(X_train)
    
    # Calculate evaluation metrics
    mae = mean_absolute_error(y_train, y_train_pred)
    mse = mean_squared_error(y_train, y_train_pred)
    r2 = r2_score(y_train, y_train_pred)
    
    # Store results
    results[name] = {'MAE': mae, 'MSE': mse, 'R-squared': r2}
    print(results)
    
# Display results
for name, result in results.items():
    print(f"Model: {name}")
    print(f"  Mean Absolute Error: {result['MAE']}")
    print(f"  Mean Squared Error: {result['MSE']}")
    print(f"  R-squared: {result['R-squared']}")
```

```python
print(results)
```

```python
from joblib import dump, load

best_model = None
best_mae = float('inf') 
best_mse = float('inf')  
best_r2 = -float('inf')  

for name, result in results.items():
    if result['MAE'] < best_mae and result['MSE'] < best_mse and result['R-squared'] > best_r2:
        best_model = name
        best_mae = result['MAE']
        best_mse = result['MSE']
        best_r2 = result['R-squared']
        print(best_model)


deployed_model = models[best_model]


deployed_model.fit(X_train, y_train)


joblib_file = "best_model.joblib"
dump(deployed_model, joblib_file)


loaded_model = load(joblib_file)
```
