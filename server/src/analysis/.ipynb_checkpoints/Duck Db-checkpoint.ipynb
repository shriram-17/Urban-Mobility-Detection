```python
import duckdb
```

```python
conn = duckdb.connect(database='your_database.db')
```

```python
conn.execute("CREATE TABLE mobility_data AS SELECT * FROM read_csv_auto('../analysis/urban_mobility_data_past_year.csv')")
```

```python
result = conn.execute("SELECT * FROM mobility_data LIMIT 0")
column_names = [desc[0] for desc in result.description]
print("Column Names:", column_names)
```

```python
result = conn.execute("SELECT * FROM mobility_data WHERE road_type = 'Highway'")
```

```python
query_1 = "SELECT COUNT(*) FROM mobility_data"

# Query 2: Average temperature by weather condition
query_2 = """
    SELECT weather_conditions, AVG(temperature) AS avg_temperature
    FROM mobility_data
    GROUP BY weather_conditions
"""

# Query 3: Sum of precipitation by road type
query_3 = """
    SELECT road_type, SUM(precipitation) AS total_precipitation
    FROM mobility_data
    GROUP BY road_type
"""

# Execute queries
result_1 = conn.execute(query_1).fetchall()
result_2 = conn.execute(query_2).fetchall()
result_3 = conn.execute(query_3).fetchall()
```

```python
print(result_1,result_2,result_3)
```

```python
result = conn.execute("SELECT * FROM mobility_data WHERE road_type = 'Highway'")
data = result.fetchdf()
```

```python
import matplotlib.pyplot as plt

# Create a histogram of temperatures
plt.hist(data['temperature'], bins=20, edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='temperature', y='humidity', data=data)
plt.title('Temperature vs. Humidity')
plt.show()
```

```python

# Query the data
result = conn.execute("SELECT vehicle_type, COUNT(*) as count FROM mobility_data GROUP BY vehicle_type")
data = result.fetchdf()

# Create a bar plot
plt.bar(data['vehicle_type'], data['count'])
plt.title('Vehicle Type Counts')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.show()
```

```python
# Query the data
result = conn.execute("SELECT weather_conditions, temperature FROM mobility_data")
data = result.fetchdf()

# Create a box plot
data.boxplot(column='temperature', by='weather_conditions')
plt.title('Temperature by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Temperature')
plt.show()
```

```python
import seaborn as sns

# Query the data
result = conn.execute("SELECT temperature, humidity, wind_speed, precipitation, population_density FROM mobility_data")
data = result.fetchdf()

# Calculate correlations
corr = data.corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Query the data
result = conn.execute("SELECT vehicle_type, COUNT(*) as count FROM mobility_data GROUP BY vehicle_type")
data = result.fetchdf()

# Create a pie chart
plt.figure(figsize=(10,6))
plt.pie(data['count'], labels=data['vehicle_type'], autopct='%1.1f%%')
plt.title('Vehicle Type Proportions')
plt.show()
```

```python
# Query the data
result = conn.execute("SELECT weather_conditions, temperature FROM mobility_data")
data = result.fetchdf()

# Create a violin plot
sns.violinplot(x='weather_conditions', y='temperature', data=data)
plt.title('Temperature Distributions by Weather Condition')
plt.show()
```

```python
# Query the data
result = conn.execute("SELECT temperature, humidity, wind_speed, precipitation FROM mobility_data")
data = result.fetchdf()

# Create a pair plot
sns.pairplot(data)
plt.show()
```
